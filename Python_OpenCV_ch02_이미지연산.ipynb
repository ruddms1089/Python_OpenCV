## OpenCV를 활용한 ROI(Region Of Interrast: 관심영역 추출)
ROI : 특정 이미지에서 어떤 유의미한 특징을 추출할 때 특징이 들어있는 공간

````python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("./data/dog.jpg")

# Numpy Slicing : ROI 처리 기능
roi = img[100:400, 250:550] # 관심영역 추출

# ROI 단위로 이미지 복사하기
img[0:300, 0:300] = roi # 해당 위치에 관심영역 넣기

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
````

##OpenCV 활용한 픽셀별 색상 다루기

````python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("./data/dog.jpg")
img [:, :, 2] = 0 # RED를 0으로 변경(모든 픽셀에서 RED 제거)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
````
